## 2. Simulate External Context (Temperature, Load)

```python
# Load dataset
df = pd.read_csv('../data/ai4i2020.csv')
print("Dataset Shape:", df.shape)
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Simulate ambient temperature (Celsius) - normal distribution
df['ambient_temp_C'] = np.random.normal(loc=28, scale=5, size=len(df))

# Simulate factory load percentage - uniform distribution
df['factory_load_pct'] = np.random.uniform(50, 100, size=len(df))

print("External context signals simulated:")
print(df[['ambient_temp_C', 'factory_load_pct']].describe())
```

## Commentary — Simulated External Context

Two external environmental signals were simulated to represent real-world V2X/IoT context data:

- **ambient_temp_C**: Simulated using Normal distribution (mean=28°C, std=5°C) — represents typical factory ambient temperature variation
- **factory_load_pct**: Simulated using Uniform distribution (50%-100%) — represents factory operational load percentage

**random_state=42** was used to ensure reproducibility — the same random values will be generated every time this notebook runs, allowing consistent results across the team.

These signals will be merged with internal sensor data and tested in the ablation study to prove external context improves prediction accuracy.
