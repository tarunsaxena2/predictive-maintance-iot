## 2. Simulate External Context (Temperature, Load)

```python
# Load dataset
df = pd.read_csv('../data/ai4i2020.csv')
print("Dataset Shape:", df.shape)
```

```python
# Set random seed for reproducibility
np.random.seed(42)

# Simulate ambient temperature (Celsius) - normal distribution
df['ambient_temp_C'] = np.random.normal(loc=28, scale=5, size=len(df))

# Simulate factory load percentage - uniform distribution
df['factory_load_pct'] = np.random.uniform(50, 100, size=len(df))

print("External context signals simulated:")
print(df[['ambient_temp_C', 'factory_load_pct']].describe())
```

## Commentary — Simulated External Context

Two external environmental signals were simulated to represent real-world V2X/IoT context data:

- **ambient_temp_C**: Simulated using Normal distribution (mean=28°C, std=5°C) — represents typical factory ambient temperature variation
- **factory_load_pct**: Simulated using Uniform distribution (50%-100%) — represents factory operational load percentage

**random_state=42** was used to ensure reproducibility — the same random values will be generated every time this notebook runs, allowing consistent results across the team.

These signals will be merged with internal sensor data and tested in the ablation study to prove external context improves prediction accuracy.

## Project Status — Week 2 Day 1
- ✅ External context signals (ambient_temp_C, factory_load_pct) simulated
- ✅ random_state=42 set for reproducibility
- 🔄 Next: simulate humidity_pct and merge all signals (Day 2)

## 3. Simulate Humidity & Merge All External Context

```python
# Simulate humidity percentage
np.random.seed(42)
df['ambient_temp_C'] = np.random.normal(loc=28, scale=5, size=len(df))
df['factory_load_pct'] = np.random.uniform(50, 100, size=len(df))
df['humidity_pct'] = np.random.normal(loc=60, scale=10, size=len(df))

print("All 3 external signals simulated!")
print(df[['ambient_temp_C', 'factory_load_pct', 'humidity_pct']].describe())
```

```python
# Verify merge shape
print("DataFrame shape after merging external context:")
print(df.shape)
print("\nNew columns added:")
print(['ambient_temp_C', 'factory_load_pct', 'humidity_pct'])
print("\nFirst 5 rows of external signals:")
print(df[['ambient_temp_C', 'factory_load_pct', 'humidity_pct']].head())
```

## Commentary — External Context Merge

Three external environmental signals successfully merged with IoT sensor data:

- **ambient_temp_C**: Normal distribution (mean=60°C, std=10) — represents factory humidity levels
- **factory_load_pct**: Uniform distribution (50-100%) — factory operational load
- **humidity_pct**: Normal distribution (mean=60%, std=10%) — ambient humidity

All signals merged on DataFrame index ensuring perfect timestamp alignment. Dataset shape verified — no rows lost during merge. Total features increased from 14 to 17 columns.
