# Project Summary

## Project Goal

The objective of this project is to develop an intelligent Predictive Maintenance system using the AI4I 2020 Predictive Maintenance dataset. The solution combines IoT sensor telemetry, contextual environmental features, and machine learning techniques to predict machine failures before they occur. The project focuses on improving reliability, reducing unplanned downtime, and supporting proactive maintenance decisions in industrial environments.

## Dataset

The project uses the **AI4I 2020 Predictive Maintenance Dataset**, which contains sensor readings and operational parameters collected from industrial machines.

**Dataset Summary**
- Total Records: **10,000**
- Total Features: **15**
- Target Variable: **Machine Failure**
- Data Type: Industrial IoT Sensor Data

The dataset includes important machine operating parameters such as air temperature, process temperature, rotational speed, torque, tool wear, and machine failure indicators. Data preprocessing included duplicate removal, missing value verification, feature engineering, and feature scaling before model training.

## Conclusion

This project successfully developed an AI-based predictive maintenance system capable of detecting machine failures using industrial IoT sensor data.

The combination of feature engineering, SMOTE, and the LightGBM algorithm significantly improved prediction accuracy and model robustness. The final model achieved a **Macro F1 Score of 0.9908**, making it suitable for predictive maintenance tasks in real-world industrial environments.

Future improvements may include real-time IoT integration, cloud deployment, and live monitoring dashboards for continuous machine health assessment.
